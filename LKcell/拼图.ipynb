```python
######拼成有细胞核的tiff
import os
from PIL import Image
from tqdm import tqdm
import glob
patch_dir = "/home/stat-huamenglei/LKCell/WSI/HighRS/1534164/patches"
files = [
    f for f in glob.glob(os.path.join(patch_dir, "*.png"))
    if f.endswith("_overlay.png") and not f.endswith("_overlay_overlay.png")
]

patches = []
for f in files:
    parts = f.split("_")
    x, y = int(parts[1]), int(parts[2].split(".")[0])
    patches.append((x, y, f))

xs = [p[0] for p in patches]
ys = [p[1] for p in patches]
min_x, max_x = min(xs), max(xs)
min_y, max_y = min(ys), max(ys)

# 用第一张 patch 确定尺寸
with Image.open(os.path.join(patch_dir, patches[0][2])) as sample:
    pw, ph = sample.size

big_img = Image.new("RGB", ((max_x - min_x + 1) * pw,
                            (max_y - min_y + 1) * ph))

# 边读边 paste
for x, y, f in tqdm(patches, desc="拼接进度"):
    with Image.open(os.path.join(patch_dir, f)) as img:
        big_img.paste(img, ((y - min_y) * pw, (x - min_x) * ph))

# 保存为 TIFF，速度比 PNG 快
big_img.save("stitched_1534164.tiff", compression=None)
print("✅ 拼接完成: stitched_1534164.tiff")
```

```python
#########用于寻找坐标的tiff

import os
from PIL import Image, ImageDraw, ImageFont
from tqdm import tqdm
import glob
patch_dir = "/home/stat-huamenglei/LKCell/WSI/HighRS/1534164/patches"
files = [f for f in glob.glob(os.path.join(patch_dir, "*.png")) 
               if not f.endswith("_overlay.png")]

patches = []
for f in files:
    parts = f.split("_")
    x, y = int(parts[1]), int(parts[2].split(".")[0])
    patches.append((x, y, f))

xs = [p[0] for p in patches]
ys = [p[1] for p in patches]
min_x, max_x = min(xs), max(xs)
min_y, max_y = min(ys), max(ys)

# 用第一张 patch 确定尺寸
with Image.open(os.path.join(patch_dir, patches[0][2])) as sample:
    pw, ph = sample.size

big_img = Image.new("RGB", ((max_x - min_x + 1) * pw,
                            (max_y - min_y + 1) * ph))

# 字体（Linux 下可能需要改路径，比如 /usr/share/fonts/...）
try:
    font = ImageFont.truetype("DejaVuSans-Bold.ttf", 20)
except:
    font = ImageFont.load_default()

# 边读边 paste
for x, y, f in tqdm(patches, desc="拼接进度"):
    with Image.open(os.path.join(patch_dir, f)) as img:
        X_big = (y - min_y) * pw
        Y_big = (x - min_x) * ph
        big_img.paste(img, (X_big, Y_big))

        # 在图上写 patch 名字（去掉扩展名）
        draw = ImageDraw.Draw(big_img)
        parts = os.path.basename(f).split("_")  # ['1534164', '51', '35']
        coords = f"{parts[-2]}_{parts[-1].split('.')[0]}"  # "51_35"
        draw.text((X_big + 5, Y_big + 5), coords, fill=(255, 0, 0), font=font)
        # draw.text((X_big + 5, Y_big + 5), f.split(".")[0], fill=(255, 0, 0), font=font)

# 保存为 TIFF
big_img.save("stitched_1534164_annotated.tiff", compression=None)
print("✅ 拼接完成: stitched_1534164_annotated.tiff")
```
